This next bit is a somewhat complicated way of turning the sequence of note on and off events into whole notes with a duration<br>
The issue is that each note has a start and end that may have other notes starting and ending arbitrarily between them

```python
note_starts = all_songs.loc[all_songs.type == 'note_on_c']
note_starts
```

```python
note_ends = all_songs.loc[all_songs.type == 'note_off_c']
note_ends
```

```python
def selection(start_note):
    checks = []
    for end_note in note_ends.values:
        check1 = (start_note[0] == end_note[0])
        check2 = (start_note[1] == end_note[1])
        check3 = (start_note[4] == end_note[4])
        checks.append(all([check1, check2, check3]))
    candidates = note_ends.loc[checks]
    
    # this gets the time difference and if it is negative it makes it arbitrarily large so that it isn't counted as closest
    differences = np.array([end_note_time - start_note[2] if ((end_note_time - start_note[2]) < 0) else np.NaN 
                            for end_note_time in candidates.time.values])
    
    ending_note = candidates.iloc[differences.argmin()]
    return ending_note
    
```

Now we make a dictionary to convert note numbers into letters

```python
note_data = sorted(all_notes.note.unique())
note_data
```

```python
note_dict = { key: 'value' for key in note_data }
note_dict
```

```python
abs(-60)
```

```python
def note_assign(number):
    # here are the notes in the scale
    note_list = ['c', 'c#', 'd', 'd#', 'e', 'f', 'f#', 'g', 'g#', 'a', 'a#', 'b' ]
    offset = number - 60
    # if the note value is > 60, offest will be positive, and modular division will work
    if offset >= 0:
        note = note_list[offset%12]
    if offset < 0:
        note = note_list[-(abs(offset)%12)]
        
    # now to assign the octave number of the note, where 60 = c4 and 56 = g#3, we need // (floor division)
    if (number-57) >= 0:
        note = note + str( 4 + ((number-57)//12) )
    if (number-57) < 0:
        note = note + str( 3 - (abs(number-57)//13) )
    
    return note
```

```python
for note in note_dict.keys():
    note_dict[note] = note_assign(note)
```

```python
translated = []
for note in all_notes.note:
    translated.append(note_dict[note])
```

# Exploration

Notes by total occurance

```python
plt.figure(figsize=(10,12))
sns.countplot(data=all_notes, y='note', color='teal',
              order=all_notes.groupby('note')['exists'].count().sort_values(ascending=False).index)
plt.title('Notes by total occurance throughout all songs')
plt.show()
```

Title and most common note

```python
common_df = all_notes.copy().groupby(['title', 'note']).count().reset_index()
most_common_notes = {
    title: common_df.loc[common_df.loc[common_df['title'] == title].exists.idxmax()].note
 for title in all_notes.title.unique() }
all_notes_common = all_notes.copy()
all_notes_common['mode'] = [ most_common_notes[title] for title in all_notes.title ]
all_notes_common = all_notes_common.loc[:,['title', 'mode']]
all_notes_common = all_notes_common.loc[[not value for value in all_notes_common.duplicated()]]
all_notes_common.reset_index(inplace=True)
all_notes_common = all_notes_common.loc[:,['title', 'mode']]
all_notes_common
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(data=all_notes_common, y='title', x='mode')
plt.title('Most common note for each song')
plt.xlabel('Most common note')
plt.show()
```
